```python
import matplotlib.pyplot as plt
```

```python
#Ravdass dataset
ravdass_train_dataset = {
    "Neutral":288,
    "Disgust":192,
    "Happy":192,
    "Sad":192,
    "Angry":192,
    "Surprise":192,
    "Fear":192
}
```

```python
plt.bar(ravdass_train_dataset.keys(),ravdass_train_dataset.values())
plt.xlabel("Emotions")
plt.ylabel("Audio Samples")
plt.title("RAVDASS DATASET")
plt.show()
```
